```python
%config IPCompleter.greedy=True
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
X = np.array([1, 3, 5])
Y = np.array([6, -2, 4])
data = np.c_[X, Y]
```

```python
plt.plot(X, Y)
plt.show()
```

```python
n = len(data) - 1
A = np.empty((0, 4 * n), dtype=np.float64)
b = []
print(A)
# edges equations
for i, (p_l, p_r)  in enumerate(zip(data, data[1:])):
    left = np.zeros((4*n), dtype=np.float64)
    left[4*i:4*i + 4] = [1, 0, 0, 0]
    A = np.vstack((A, left))
    b.append(p_l[1])
    
    h = p_r[0] - p_l[0]
    right = np.zeros((4*n), dtype=np.float64)
    right[4*i:4*i + 4] = [h**i for i in range(4)]
    A = np.vstack((A, right))
    b.append(p_r[1])
   
    print(A)
    print(b)
```
